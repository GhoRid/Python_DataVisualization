# 15. 산점도 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Window
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트를 사용할 때 마이너스 글자가 깨지는 걸 방지
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s = sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500  # 1학년 500, 2학년 1000, 3학년 1500
plt.scatter(df['영어'], df['수학'], s = sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.scatter(df['영어'], df['수학'], s = sizes, c = df['학년'], cmap = 'viridis', alpha = 0.3)    # c : 분포 별 색깔 자동 정의
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(df['영어'], df['수학'], s = sizes, c = df['학년'], cmap = 'viridis', alpha = 0.3)    # c : 분포 별 색깔 자동 정의
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

plt.colorbar( ticks = [1,2 ,3], label = '학년', shrink = 0.5, orientation = 'horizontal')   
# shrink : colorbar 크기  orientation = 'horizontal' : color bar 위치 아래로
```
