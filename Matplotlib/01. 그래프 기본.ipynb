# Matplotlib
다양한 형태의 그래프를 통해서 데이터를 시각화 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
```

```python
print(plt.plot(x, y))
```

```python
plt.plot(x, y)
plt.show()
```

## Title 설정

```python
plt.plot(x, y)
plt.title('Line Gragh')
```

```python
plt.plot(x, y)
plt.title('꺾은 선 그래프')
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Window
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트를 사용할 때 마이너스 글자가 깨지는 걸 방지
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 목록 확인
[f.name for f in fm.fontManager.ttflist]     # 폰트매니저에 있는 폰트를 하나씩 '이름만' 출력 (f.name)
```
