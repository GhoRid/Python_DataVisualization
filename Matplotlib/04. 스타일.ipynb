# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Window
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트를 사용할 때 마이너스 글자가 깨지는 걸 방지
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth = 5 )   # 라인 두께 설정
```

## 마커 (marker)
  
matplotlib markers 검색하기

```python
plt.plot(x, y, marker = 'o')  # 데이터 위치에 점 표시
```

```python
plt.plot(x, y, marker = 'o', linestyle = 'None')   # linestyle = 'None': 선 지우기
```

```python
plt.plot(x, y, marker = 'v')   
```

```python
plt.plot(x, y, marker = 'v', markersize = 10)   
```

```python
plt.plot(x, y, marker = 'x', markersize = 10)  # 대문자 x도 가능 
```

```python
plt.plot(x, y, marker = 'X', markersize = 10, markeredgecolor = 'red')   #엣지 컬러 : 외곽선 색
```

```python
plt.plot(x, y, marker = 'o', markersize = 20, markeredgecolor = 'red', markerfacecolor = 'yellow') # 마커 색깔
```

## 선 스타일
  
matplotlib linestyle 검색하기

```python
plt.plot(x, y, linestyle= ':')
```

```python
plt.plot(x, y, linestyle= '--')
```

```python
plt.plot(x, y, linestyle= '-.')
```

```python
plt.plot(x, y, linestyle= '-')
```

## 색깔

```python
plt.plot(x, y, color = 'pink')
```

## 포맷
간단하게 라인 모습 변경

```python
plt.plot(x, y, 'ro--')   # color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go')
```

## 축약어

```python
plt.plot(x, y, marker = 'o', mfc = 'red', ms = '10', mec = 'blue', ls = ':') 
# mfc = markerfacecolor          ms = markerstyle            mec = marker edge color         ls = linestyle
```

## 투명도

```python
plt.plot(x, y, marker = 'o', mfc = 'red', ms = '10', mec = 'blue', ls = ':', alpha = 0.3)  # alpha : 투명도
```

## 그래프 크기

```python
plt.figure(figsize = (10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize = (5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize = (10, 5), dpi= 200)    # dpi = 해상도
plt.plot(x, y)
```

```python
plt.figure(figsize = (10, 5), dpi= 50)    # dpi =  dots per incnh  해상도
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor = 'yellow')
plt.plot(x, y)
```
