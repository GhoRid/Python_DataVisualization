# 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Window
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트를 사용할 때 마이너스 글자가 깨지는 걸 방지
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.title('꺾은 선 그래프', fontdict = {'family': 'HYGungSo-Bold', 'size': '20'}) # 개별 폰트 설정
```

```python
plt.plot(x, y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
plt.plot(x, y)
plt.xlabel('X축', color = 'red')    # 개별 색 설정 가능
plt.ylabel('Y축', color = '#00aa00')
```

```python
plt.plot(x, y)
plt.xlabel('X축', color = 'red', loc = 'right')    # left, center, right 세 가지 옵션   loc = location
plt.ylabel('Y축', color = '#00aa00')               # top, center, bottom
```

```python
plt.plot(x, y)
plt.xticks([1,2,3]) # X 축에서 나타낼 값들을 설정할 수 있음
plt.yticks([3,6,9,12]) # Y 축에서 나타낼 값들을 설정할 수 있음
plt.show()
```
