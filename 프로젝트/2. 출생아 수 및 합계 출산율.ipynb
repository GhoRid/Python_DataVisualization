```python
import pandas as pd
df = pd.read_excel('142801_20230212205220982_excel.xlsx', skiprows = 2, nrows = 2, index_col = 0)
df
```

```python
df = df.T        # 행과 열을 바꾸어줌. 전치 행렬이 Transposed matrix 였다는 걸 생각하면 기억하기 쉽다.
df
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Window
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트를 사용할 때 마이너스 글자가 깨지는 걸 방지
```

```python
plt.plot(df.index, df['출생아 수'], label = '출생아 수')
plt.plot(df.index, df['합계 출산율'], label = '합계 출산율')
plt.legend()
```

```python
fig, ax1 = plt.subplots(figsize = (10, 7))
ax1.plot(df.index, df['출생아 수'], color = '#ff812d')

ax2 = ax1.twinx()  # X축을 공유하는 쌍둥이 axis 생성
ax2.plot(df.index , df['합계 출산율'], color = '#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize = (10, 7))
ax1.set_ylabel('출생아 수 (천 명)')     #subplot을 사용할 때는 set_ 를 포함해서 설정해야함
ax1.bar(df.index, df['출생아 수'], color = '#ff812d')


ax2 = ax1.twinx()  # X축을 공유하는 쌍둥이 axis 생성
ax2.set_ylabel('합계 출산율 (가입 여성 1명 당 명)')
ax2.plot(df.index , df['합계 출산율'], color = '#ffd100')
```
